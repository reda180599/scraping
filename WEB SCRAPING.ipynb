```python
#import libararies

#lxml


import requests
import lxml
from bs4 import BeautifulSoup as bss 
import csv




import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



```

```python


url = requests.get('https://www.amazon.eg/-/en/%D9%85%D9%88%D8%A8%D8%A7%D9%8A%D9%84%D8%A7%D8%AA/b/?ie=UTF8&node=21832883031&ref_=nav_cs_mobiles')


```

```python


print(url)
```

```python
#html text (page content)

print(url.text)
```

```python

soup = bss(url.text, 'lxml')

product_title = soup.find_all('span', {'class' :'a-size-base-plus'})

product_title
```

```python
print(product_title)


```

```python


product_price = soup.find_all('span', {'class' :'a-offscreen'})

product_price
```

```python



print(product_price)
```

```python
#extract the information to .csv file


products = soup.find_all('div',{'class':'_octopus-search-result-card_style_apbSearchResultItem__2-mx4'})


with open('products.csv', 'w',encoding='utf-8', newline='') as file:              # with for open and coloe the connection
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Product Price'])
    
    
    for product in products:
        title = soup.find('span', {'class' :'a-size-base-plus'}).text
        price_p = soup.find('span', {'class' :'a-offscreen'}).text
        
        
        
        
        writer.writerow([title, price_p])
    
    

```

```python
print("done")
```
